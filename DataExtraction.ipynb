```python
import os
import json
import pandas as pd
from sqlalchemy import create_engine
```

```python
csv_filepath = "csv/"
```

```python

# create SQLite database connection
engine = create_engine("sqlite:///phonepay.db")
```

### 1. Aggregated

#### 1.1 Aggregated - Insurance

```python
# 
# pulse\data\aggregated\insurance\country\india\state

# Path to dataset
path = "pulse/data/aggregated/insurance/country/india/state/"

data = []

# Loop through states
for state in os.listdir(path):
    state_path = path + state + "/"
    
    for year in os.listdir(state_path):
        year_path = state_path + year + "/"
        
        for file in os.listdir(year_path):
            file_path = year_path + file
            
            with open(file_path, 'r') as f:
                d = json.load(f)
                
                for i in d['data']['transactionData']:
                    name = i['name']
                    
                    for j in i['paymentInstruments']:
                        data.append({
                            "state": state,
                            "year": int(year),
                            "quarter": int(file.strip('.json')),
                            "transaction_type": name,
                            "count": j['count'],
                            "amount": j['amount']
                        })

df = pd.DataFrame(data)

df.to_csv("csv/aggreated_insurance.csv",index=False)

# Inserting Data into sql

table_name = 'aggreated_insurance'
df.to_sql(name=table_name, con=engine, if_exists='replace', index=False)

print(f"Data inserted successfully to {table_name} table")
df.head()
```

#### 1.3 Aggregated - user

```python

# Path to dataset
path = "pulse/data/aggregated/user/country/india/state/"

data = []

# Loop through states
for state in os.listdir(path):
    state_path = path + state + "/"
    
    for year in os.listdir(state_path):
        year_path = state_path + year + "/"
        
        for file in os.listdir(year_path):
            file_path = year_path + file
            
            with open(file_path, 'r') as f:
                d = json.load(f)
                if d['data']['usersByDevice']:
                    for i in d['data']['usersByDevice']:
                        # {"brand":"Vivo","count":15056,"percentage":0.2424671873741847}
                        data.append({
                            "state": state,
                            "year": int(year),
                            "quarter": int(file.strip('.json')),
                            "brand": i['brand'],
                            "count": i['count'],
                            "amount": i['percentage']
                        })

df = pd.DataFrame(data)

df.to_csv(csv_filepath+"aggreated_user.csv",index=False)

# Inserting Data into sql for analysing
table_name = 'aggreated_user'
df.to_sql(name=table_name, con=engine, if_exists='replace', index=False)

print(f"Data inserted successfully to {table_name} table")
df.head()
```

### 2. Map

#### 2.1 Map - insurance

```python
# pulse\data\map\insurance\hover\country\india\state

# Path to dataset
path = "pulse/data/map/insurance/hover/country/india/state/"

data = []

# Loop through states
for state in os.listdir(path):
    state_path = path + state + "/"
    
    for year in os.listdir(state_path):
        year_path = state_path + year + "/"
        
        for file in os.listdir(year_path):
            file_path = year_path + file
            
            with open(file_path, 'r') as f:
                d = json.load(f)
                
                if d["data"] and "hoverDataList" in d["data"]:
                    for  item in d["data"]["hoverDataList"]:
                        name=item["name"]
                        count=item["metric"][0]["count"]
                        amount=item["metric"][0]["amount"]                        
                        
                        data.append({
                            "state":state,
                            "year": int(year),
                            "quarter":int(file.strip('.json')),
                            "distric":name,
                            "insurance_count":count,
                            "insurance_amount":amount
                        })

df = pd.DataFrame(data)
# Storing data to csv file
df.to_csv(csv_filepath+"map_insurance.csv",index=False)

# Inserting Data into sql for analysing
table_name = 'map_insurance'
df.to_sql(name=table_name, con=engine, if_exists='replace', index=False)

print(f"Data inserted successfully to {table_name} table")
df.head()
```

#### 2.3 Map - user

```python

# Path to dataset
path = "pulse/data/map/user/hover/country/india/state/"

data = []

# Loop through states
for state in os.listdir(path):
    state_path = path + state + "/"
    
    for year in os.listdir(state_path):
        year_path = state_path + year + "/"
        
        for file in os.listdir(year_path):
            file_path = year_path + file
            
            with open(file_path, 'r') as f:
                d = json.load(f)

                hover_data = d.get("data",{}).get("hoverData",{})
            
                for district_name,metrics in hover_data.items():
                    registered_users = metrics.get("registeredUsers",0)
                    app_opens = metrics.get("appOpens",0)
                    
                    data.append({
                        "state": state,
                        "year": int(year),
                        "quarter": int(file.strip('.json')),
                        "district":district_name.title(),
                        "registered_users":registered_users,
                        "app_opens":app_opens
                    })

df = pd.DataFrame(data)
# Storing data into csv file
df.to_csv(csv_filepath+"map_user.csv",index=False)

# Inserting Data into sql for analysing
table_name = 'map_user'
df.to_sql(name=table_name, con=engine, if_exists='replace', index=False)

print(f"Data inserted successfully to {table_name} table")
df.head()
```

### 3. Top

#### 3.1 Top - insurance

```python
# Path to dataset
path = "pulse/data/top/insurance/country/india/state/"

districts = []
pincodes = []

# Loop through states
for state in os.listdir(path):
    state_path = path + state + "/"
    
    for year in os.listdir(state_path):
        year_path = state_path + year + "/"
        
        for file in os.listdir(year_path):
            file_path = year_path + file
            
            with open(file_path, 'r') as f:
                data = json.load(f)
                
                if data.get("data"):
                    # Adding districts data
                    for d in data["data"]["districts"]:
                        entity_name_district=d["entityName"]
                        count_districts=d["metric"]["count"]
                        amount_district=d["metric"]["amount"]
                        
                        districts.append({
                        "state": state,
                        "year": int(year),
                        "quater": int(file.strip('.json')),
                        "district_entity_name":entity_name_district,
                        "district_count":count_districts,
                        "district_amount":amount_district})        
                
                    # Adding pincodes data 
                    for p in data["data"]["pincodes"]:
                        entity_name_pincodes=p["entityName"]   
                        count_pincode=p["metric"]["count"]
                        amount_pincode=p["metric"]["amount"]
                    
                        pincodes.append({
                        "state": state,
                        "year": int(year),
                        "quater": int(file.strip('.json')),
                        "pincode_entity_name":entity_name_pincodes,
                        "pincode_count":count_pincode,
                        "pincode_amount":amount_pincode})              

top_insurance_dist_df = pd.DataFrame(districts)
top_insurance_pincode_df = pd.DataFrame(pincodes)

# Store the data into csv file
top_insurance_dist_df.to_csv(csv_filepath+"top_insurance_districs.csv",index=False)

# Inserting Data into sql for analysing
table_name = 'top_insurance_districs'
top_insurance_dist_df.to_sql(name=table_name, con=engine, if_exists='replace', index=False)

print(f"Data inserted successfully to {table_name} table")

top_insurance_pincode_df.to_csv(csv_filepath+"top_insurance_pincodes.csv",index=False)

# Inserting Data into sql for analysing
table_name = 'top_insurance_pincodes'
top_insurance_pincode_df.to_sql(name=table_name, con=engine, if_exists='replace', index=False)

print(f"Data inserted successfully to {table_name} table")
```

#### 3.3 Top - User

```python
# Path to dataset
path = "pulse/data/top/user/country/india/state/"

districts = []
pincodes = []

# Loop through states
for state in os.listdir(path):
    state_path = path + state + "/"
    
    for year in os.listdir(state_path):
        year_path = state_path + year + "/"
        
        for file in os.listdir(year_path):
            file_path = year_path + file
            
            with open(file_path, 'r') as f:
                data = json.load(f)
                
                if data.get("data"):
                    # Adding districts data
                    for d in data["data"]["districts"]:
                        district=d["name"]
                        reg_users=d["registeredUsers"]

                        districts.append({
                        "state": state,
                        "year": int(year),
                        "quater": int(file.strip('.json')),
                        "district_name":district,
                        "registeredUsers":reg_users})        
                
                    # Adding pincodes data 
                    for p in data["data"]["pincodes"]:
                        pincode=p["name"]   
                        reg_users=p["registeredUsers"]
                    
                        pincodes.append({
                        "state": state,
                        "year": int(year),
                        "quater": int(file.strip('.json')),
                        "pincode":pincode,
                        "registeredUsers":reg_users})              


##########################
#  Districts data
##########################

top_user_dist_df = pd.DataFrame(districts)
top_user_dist_df.to_csv(csv_filepath+"top_user_districs.csv",index=False)

# Inserting Data into sql for analysing
table_name = 'top_user_districs'
top_user_dist_df.to_sql(name=table_name, con=engine, if_exists='replace', index=False)

print(f"Data inserted successfully to {table_name} table")

##########################
#  Pincodes data
##########################

top_user_pincode_df = pd.DataFrame(pincodes)
top_user_pincode_df.to_csv(csv_filepath+"top_user_pincodes.csv",index=False)

# Inserting Data into sql for analysing
table_name = 'top_user_pincodes'
top_user_pincode_df.to_sql(name=table_name, con=engine, if_exists='replace', index=False)

print(f"Data inserted successfully to {table_name} table")
```
